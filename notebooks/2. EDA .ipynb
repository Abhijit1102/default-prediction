## 2. Checking weather Data is Balance or not 

```python
print(f"Default Datasets shape: {df_copy[df_copy['default'] == 1].shape}")
print(f"Not Default Datasets shape: {df_copy[df_copy['default'] == 0].shape}")
```

## 3. Satistical Test for signifcance of unbalanced data

```python
import scipy.stats as stats

# Create a contingency table
contingency_table = pd.crosstab(df_copy['default'], columns='count')

# Perform the chi-squared test
chi2, p_value, _, _ = stats.chi2_contingency(contingency_table)

# Print the test results
print("Chi-squared test results:")
print(f"Chi-squared statistic: {chi2}")
print(f"P-value: {p_value}")
```

## Conclusion

- `p-value` is 1.0. This indicates that there is no significant difference between the observed and expected frequencies in each class, suggesting that the dataset is balanced with respect to the "default" variable.

- A `p-value` of 1.0 suggests that the observed frequencies in the dataset are perfectly in line with the expected frequencies under the assumption of balance. Therefore, you fail to reject the null hypothesis, indicating that there is no evidence of significant imbalance in the dataset

## 4.  Impact of Age in Default 

```python
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.boxplot(x='default', y='AGE', data=df)
plt.title('Age Distribution by Default')
plt.xlabel('Default')
plt.ylabel('Age')
plt.show()

# Density plot
plt.figure(figsize=(8, 6))
sns.kdeplot(df_copy[df_copy['default'] == 0]['AGE'], label='Not Default')
sns.kdeplot(df_copy[df_copy['default'] == 1]['AGE'], label='Default')
plt.title('Age Distribution by Default')
plt.xlabel('Age')
plt.ylabel('Density')
plt.legend()
plt.show()

# T-test
from scipy.stats import ttest_ind

age_default = df_copy[df_copy['default'] == 1]['AGE']
age_not_default = df_copy[df_copy['default'] == 0]['AGE']

t_stat, p_value = ttest_ind(age_default, age_not_default)
print("T-test results:")
print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")
```

## Conclusion
- Based on the `t-test` result, the t-statistic is approximately 2.406 and the p-value is approximately 0.016.

- The t-test evaluates the difference in means between the age of the default and not default groups. In this case, the results indicate that there is a statistically significant difference in mean age between the two groups.

- With a `p-value` of `0.016`, which is below the common significance level of `0.05`, we can reject the null hypothesis and conclude that there is evidence to support the claim that the mean age differs significantly between the default and not default groups.

- Based on this analysis, age appears to have an impact on default. However, it's important to note that statistical significance does not necessarily imply practical significance. The effect size and the context of the problem should also be considered when interpreting the results.
